```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('nausica.jpg')
cv2.imshow('head', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#q1. convert to grayscale. 

img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
x=height, width = img_grey.shape
print(x)
img_1d=img_grey.flatten()
print(img_1d)
intensity=img_1d[2]
print(intensity)
plt.hist(img_1d, bins=100)
```

```python
cv2.imshow('grey', img_grey )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#q2 
img1=img_grey.copy()
x,y=180,140
n=100
sub_img = img1[y:y+n, x:x+n]
#out_img = []
#increase brightness
#print(sub_img)
ht, wt = sub_img.shape
out_img = np.zeros((ht, wt), dtype=np.uint8) # I needed this data frame

for i in range(ht):
    for j in range(wt):
        pixel_int=int(sub_img[i,j])
        pixel_int=pixel_int+50
        if pixel_int > 255:
            pixel_int=255
        #store
        out_img[i,j] = pixel_int 

img1[y:y+n, x:x+n] = out_img
```

```python
cv2.imshow('grey', out_img )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
